## 1 Import Library
Pada bagian ini saya mengimpor library yang akan digunakan untuk mengolah data, membuat grafik, dan mengakses dataset bawaan.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

## 2 Analisis Array Menggunakan NumPy
Saya membuat 20 angka acak lalu menghitung rata-rata, median, dan standar deviasi untuk melihat penyebaran data.

```python
data = np.random.randint(0, 100, 20)
print("Array:", data)

mean = np.mean(data)
median = np.median(data)
std = np.std(data)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
```

### Output Perhitungan

Berdasarkan hasil program, diperoleh array berisi 20 angka yaitu 34, 21, 43, 73, 4, 26, 53, 41, 27, 85, 11, 81, 34, 79, 85, 63, 39, 41, 47, dan 21. Nilai rata-rata dari data tersebut adalah 45.4, median sebesar 41.0, dan standar deviasi sebesar 24.30308622377002. Hal ini menunjukkan bahwa data memiliki rata-rata sekitar 45 dengan penyebaran nilai yang cukup beragam.

## 3 Membuat Data Nilai Mahasiswa
Saya membuat tabel berisi 10 mahasiswa dengan nilai acak kemudian mencari 3 nilai tertinggi.

```python
nama = ["Virr","Elle","Arru","Jade","Tropis","Zhe","Wawan","Allie","Petek","GYJ"]
nilai = np.random.randint(50, 100, 10)

df = pd.DataFrame({
    "Nama": nama,
    "Nilai": nilai
})

print("Data Nilai Mahasiswa:")
print(df)

top3 = df.sort_values(by="Nilai", ascending=False).head(3)

print("3 Nilai Tertinggi:")
print(top3)
```

### Output Data Nilai Mahasiswa

Data nilai yang diperoleh adalah sebagai berikut:
Virr memperoleh nilai 89, Elle 92, Arru 85, Jade 96, Tropis 90, Zhe 67, Wawan 56, Allie 81, Petek 97, dan GYJ 57.

Setelah diurutkan dari nilai tertinggi, tiga mahasiswa dengan nilai terbaik adalah Petek dengan nilai 97, Jade dengan nilai 96, dan Elle dengan nilai 92.

Dari hasil tersebut dapat dilihat bahwa Petek memiliki nilai paling tinggi di antara seluruh mahasiswa.

## 4 Visualisasi Data
Data nilai divisualisasikan menggunakan grafik batang dan grafik garis agar lebih mudah dianalisis.

```python
plt.figure()
plt.bar(df["Nama"], df["Nilai"])
plt.title("Grafik Bar Nilai Mahasiswa")
plt.xlabel("Nama")
plt.ylabel("Nilai")
plt.show()

plt.figure()
plt.plot(df["Nama"], df["Nilai"])
plt.title("Grafik Line Nilai Mahasiswa")
plt.xlabel("Nama")
plt.ylabel("Nilai")
plt.show()
```

## Penjelasan Output

Dari grafik yang dihasilkan, dapat dilihat perbedaan nilai masing-masing mahasiswa. Grafik batang menunjukkan perbandingan nilai secara jelas, sedangkan grafik garis memperlihatkan pola kenaikan dan penurunan nilai berdasarkan urutan nama.

## 5 Menggunakan Dataset Wine
Saya memuat dataset wine dari scikit-learn untuk melihat jumlah data, fitur, dan kelas yang tersedia.

```python
wine = load_wine()

df_wine = pd.DataFrame(wine.data, columns=wine.feature_names)
df_wine["target"] = wine.target

print("Jumlah Data dan Fitur:", df_wine.shape)
print("Nama Kelas:", wine.target_names)
print(df_wine.head())
```

## Penjelasan Output

Dari hasil yang ditampilkan, dataset terdiri dari 178 data dan 14 fitur. Data ini memiliki tiga kelas berbeda. Setiap baris berisi nilai dari beberapa fitur kimia, sedangkan kolom target menunjukkan kelas dari masing-masing data.
